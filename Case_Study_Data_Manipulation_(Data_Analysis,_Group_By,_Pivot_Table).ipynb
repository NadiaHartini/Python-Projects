## **Soal 4**

**Analisis Zona Penjualan**

Anda ingin menganalisis jumlah total penjualan properti di setiap zona (MSZoning) berdasarkan tahun penjualan (YrSold). Gunakan tabel silang (crosstab) untuk menghitung jumlah penjualan di setiap zona untuk tiap tahun. Setelah itu, tambahkan kolom baru yang menandai zona dengan penjualan tinggi, di mana zona tersebut memiliki total penjualan di atas rata-rata.


```python
#Membuat crosstab untuk menghitung jumlah penjualan di setiap zona untuk tiap tahun
tabel_analisis_zona_penjualan = pd.crosstab(index=df["YrSold"], columns=df["MSZoning"], values=df["SalePrice"], aggfunc='sum')
#Menyimpan nilai rata - rata per tahun dalam variabel rata - rata per tahun
rata_rata_per_tahun = tabel_analisis_zona_penjualan.mean(axis = 1)

#Membuat fungsi untuk mengidentifikasi zona dengan penjualan tinggi (zona tersebut memiliki total penjualan di atas rata - rata)
def diatas_rata_rata(row):
  for zone, sales in row.items():
    if sales > rata_rata_per_tahun[row.name]:
      return zone

#Membuat kolom baru dengan nama kolom Zona Penjualan Tinggi (di atas rata-rata) dimana nilai entri data adalah hasil fungsi pengidentidaksi zona dengan penjualan tinggi terhadap data
tabel_analisis_zona_penjualan["Zona Penjualan Tinggi (di atas rata-rata)"] = tabel_analisis_zona_penjualan.apply(diatas_rata_rata, axis = 1)

#Menampilkan hasil tahap akhir
print(tabel_analisis_zona_penjualan)
```

**Tabel pivot MSZoning dan SalePrice untuk tahun yang berbeda (Dari 2006 sampai 2010) memberikan beberapa informasi, yakni sebagai berikut:**

1. **Analisis Zona Penjualan Tinggi (di atas rata - rata)**

* Tabel menunjukkan rata-rata harga jual untuk zona C (semua), FV, RH, RL, dan RM dari tahun 2006 samapai 2010. Walaupun rata-rata harga jual bervariasi, tetapi properti pada Zona RL selalu memiliki harga tertinggi. Hal ini berarti properti dalam klasifikasi zona ini dijual dengan harga lebih tinggi daripada yang lain.

2. **Analisis YrSold**

* Sebagian besar jenis tanah pada tahun 2010 dijual dengan harga yang lebih rendah dari 2006 hingga 2009. Tren rata - rata harga jual kemudian cenderung stabil atau berfluktuasi pada 2010.

3. **Missing Values**:

* Zonasi C (semua) dan RH tidak memiliki nilai rata - rata pada 2009 (untuk Zonasi C) dan 2010 (untuk Zonasi RH). Hal ini mungkin dikarenakan tidak adanya kegiatan perdagangan pada tahun 2009 pada Zonasi C atau 2010 pada Zonasi RH. Maka dari itu, hal ini juga mengindikasikan bahwa Zonasi C dan Zonasi RH mungkin memiliki sedikit properti untuk dijual atau tidak mampu untuk menjual properti yang dimilikinya saat itu, sehingga mengindikasikan ketidakstabilan pasar atau perubahan minat pembeli.
